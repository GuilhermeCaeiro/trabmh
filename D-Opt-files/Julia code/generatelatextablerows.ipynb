```python
files = ["permutation_instances_180-x_a_240-x.csv", "permutation_instances_280-x.csv", "permutation_instances_40-x_a_140-x.csv"]
```

```python
results = None

for file in files:
    data = pd.read_csv(file, sep=";")
    
    if results is None:
        results = data
    else: results = pd.concat([results, data])
```

```python
results_heuristics = pd.read_csv("results_heuristics.csv", sep=",", names=["n", "m", "s", "i", "solution", "time"])
```

```python
results_heuristics.head(4)
```

```python
results_heuristics["inst_name"] = results_heuristics.apply(lambda row: str(int(row.n)) + "-" + str(int(row.i)), axis=1)
```

```python
results_heuristics.head(4)
```

```python
known_solutions_and_times = {}

for index, row in results_heuristics.iterrows():
    known_solutions_and_times[row["inst_name"]] = {
        "solution": row["solution"],
        "time": row["time"],
    }
```

```python
known_solutions_and_times
```

```python
optimal_solutions = pd.read_csv("optimal_solutions_bb.csv", sep=";")
optimal_solutions["inst_name"] = optimal_solutions.apply(lambda row: str(int(row.n)) + "-" + str(int(row.i)), axis=1)
optimal_solutions
```

```python
optimal_solutions_and_times = {}

for index, row in optimal_solutions.iterrows():
    optimal_solutions_and_times[row["inst_name"]] = {
        "solution": row["z*"],
        "time": row["time (sec)"],
    }
optimal_solutions_and_times
```

```python
results["inst_name"] = results.apply(lambda row: row.instance.replace("Instance_", "").replace(".mat", "").replace("_", "-"), axis=1)
results["gap_heur"] = results.apply(lambda row: (row.best_solution_found - known_solutions_and_times[row.inst_name]["solution"]) / abs(known_solutions_and_times[row.inst_name]["solution"]), axis=1)
results["time_heur"] = results.apply(lambda row: known_solutions_and_times[row.inst_name]["time"], axis=1)
results["gap_t_heur"] = results.apply(lambda row: (row.total_time_ms - known_solutions_and_times[row.inst_name]["time"]) / abs(known_solutions_and_times[row.inst_name]["time"]), axis=1)

results["gap_bb"] = results.apply(lambda row: (row.best_solution_found - optimal_solutions_and_times[row.inst_name]["solution"]) / abs(optimal_solutions_and_times[row.inst_name]["solution"]) if row.inst_name in optimal_solutions_and_times else np.NaN, axis=1)
results["time_bb"] = results.apply(lambda row: optimal_solutions_and_times[row.inst_name]["time"] if row.inst_name in optimal_solutions_and_times else np.NaN, axis=1)
results["solution_bb"] = results.apply(lambda row: optimal_solutions_and_times[row.inst_name]["solution"] if row.inst_name in optimal_solutions_and_times else np.NaN, axis=1)
results["gap_t_bb"] = results.apply(lambda row: (row.total_time_ms - optimal_solutions_and_times[row.inst_name]["time"]) / abs(optimal_solutions_and_times[row.inst_name]["time"]) if row.inst_name in optimal_solutions_and_times else np.NaN, axis=1)
```

```python
results.head(2)
```

```python
agg = results.groupby("inst_name").aggregate({
    "best_known_result": ["max"],
    "best_solution_found": ["max", "min", "mean", "std"],
    "gap_heur": ["max"],
    "total_time_ms": ["mean"]
}).reset_index()

agg
```

```python
agg.columns = ['_'.join(col) for col in agg.columns.values]
agg
```

```python
agg.best_known_result_max[agg.inst_name_ == "40-1"]
```
